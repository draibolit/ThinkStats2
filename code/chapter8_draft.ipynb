```python
# Simulation sampling of exp dist
m = 1000
L_array = []
for i in range(1000):
    ex = np.random.exponential(2, 10000)
    L = 1/ np.mean(ex)
    L_array.append(L)

Lhat = np.array(L_array).mean()
```

```python
# Plot std vs m

def std_vs_n(n):
    ex = np.random.exponential(2, n)
#     std = np.std(ex)
    std = np.std(ex, ddof=1)
    return std

def MSE(estimate, real):
    mse = (estimate - real) ** 2
    return mse

std_array = []
n_array = []
# for n in range(10, 1000, 10):
for n in range(10, 2000, 10):
    std = std_vs_n(n)
    std_array.append(std)
    n_array.append(n)
```

```python
# sns.lineplot(m_array, std_array)
data = np.array(std_array)
index = np.array(n_array)
   
# creating the dataframe 
df = pd.DataFrame(data = data,  
                  index = index) 


sns.lineplot(data=df)
```
